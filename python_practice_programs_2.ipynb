```python
# 1.Unique Words in a sentence 

# Asking user to enter a sentence
sentence = input("Enter sentence: ")

# Splitting the sentence into individual words
words = sentence.split()

# Creating an empty set to store unique words
unique_words = set()

# Adding words one by one into the set
for word in words:
    unique_words.add(word)

# Result
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
#2.Highest Salary
# Dictionary storing employee names and salaries
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Taking the first employee as initial highest
highest_employee = ""
highest_salary = 0

# Looping through dictionary
for name in employees:
    
# Getting salary of current employee
    salary = employees[name]
    
# Comparing salaries
    if salary > highest_salary:
        highest_salary = salary
        highest_employee = name

# Displaying final result
print("Highest Salary:", highest_employee, "-", highest_salary)
```

```python
#3.Count Vowels in a String
text = input("Enter a string: ")

vowels = ['a', 'e', 'i', 'o', 'u']
count = 0

for letter in text.lower():
    if letter in vowels:
        count += 1

print("Total vowels:", count)
```

```python
#4.Count Products Above a Price Threshold
# List of product prices
prices = [450, 1200, 899, 1500, 300]

# Variable to store count
count = 0

# Loop through each price
for price in prices:
    
    # Check if price is greater than 1000
    if price > 1000:
        count += 1

# result
print("Products above 1000:", count)
```

```python
#5.Calculate Attendance Percentage
# Attendance record
attendance = ["P", "P", "A", "P", "P"]

# Count present days
present_count = 0

for status in attendance:
    
    if status == "P":
        present_count += 1

# Calculate percentage
total_days = len(attendance)
percentage = (present_count / total_days) * 100

#result
print("Attendance Percentage:", percentage)
```

```python
#6.Remove Duplicate Phone Numbers

# List of phone numbers
phone_numbers = [9876543210, 9123456789, 9876543210]

# Convert list to set
unique_numbers = set(phone_numbers)

# result
print("Unique phone numbers:", unique_numbers)
```

```python
#7.Count Character Frequency
# Input string
text = "python"

# Dictionary to store frequency
frequency = {}

# Loop through characters
for char in text:
    
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

# result
print(frequency)
```

```python
#8.Convert List to Tuple
# List of values
numbers = [10, 20, 30]

# Convert to tuple
numbers_tuple = tuple(numbers)

# result
print("Tuple:", numbers_tuple)
```

```python
#9.Check if Key Exists in Dictionary

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Key to check
key_name = "Ravi"

if key_name in employees:
    print("Employee exists")
else:
    print("Employee not found")
```

```python
#10.Calculate Average Marks
# List of marks
marks = [70, 85, 65, 90]

# Calculate total
total = 0

for mark in marks:
    total += mark

# Calculate average
average = total / len(marks)

# result
print("Average Marks:", average)
```
